For the Wine Quality Data Set, convert all the values in the quality attribute to
0 (bad) if the value is less than or equal to 6 and to 1 (good) otherwise. Normalize
all the other attributes between 0 and 1 by min-max scaling. Mention why we use
min-max scaling.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('WineQT.csv')
print(data.shape)
cols = data.columns.tolist()
print(cols)
data.head()
```

There are 1143 rows and 13 columns

['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality', 'Id']

```python
for i in range(len(data)):
    if (data.loc[i, 'quality'] <= 6):
        data.loc[i, 'quality'] = 0
    else:
        data.loc[i, 'quality'] = 1
data.head()
```

```python
min_max = data.describe().transpose()[["min", "max"]]
min_max
```

Using the above data set, we can get the min and max value for each column as


> min_max.iloc[index,0 if(min) 1 if(max)




```python
min_max.iloc[0,0]
```

In Min-Max scaling:

The formula used is:

$$x_{scaled} = \frac{x_i - x_{min}}{x_{max} -x_{min}}$$

```python
#We will consider all the columns except quality and id
cols = cols[0:-2]
for i in range(len(cols)):
  data[cols[i]] = (data[cols[i]] - min_max.iloc[i,0])/(min_max.iloc[i,1] - min_max.iloc[i,0])
data.head()
```

```python
data.describe().transpose()[["min","max"]]
```

We can see after aplying min_max scaling all the data values are there in the range of 0 and 1, quality is either 0 and 1 based on the condition defined above. and we didnt modified Id as it is not much of our use.
